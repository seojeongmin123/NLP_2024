```python
#@markdown ### ğŸ¹ <font color = 'red'>**Student Activity 1** â¤µï¸

#@markdown ğŸ‘€ğŸ¾ **1. Simple way to read and save the content of the url of your chice.**

#@markdown **e.g., Print hyper text markup language (html)**

from urllib.request import urlopen

url = 'https://google.com/'

with urlopen(url) as f:
  content = f.read().decode()
  with open('google_html', 'w') as cf:
    cf.write(content)
```
